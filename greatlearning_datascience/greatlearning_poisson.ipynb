```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
rate = 6
```

```python
n = np.arange(0,20)
```

```python
poisson = stats.poisson.pmf(n, rate)
poisson
```

```python
poisson[3] - poisson[2]
```

```python
plt.plot(n, poisson,'o-')
plt.title('Poisson: $\lambda$ = %i' % rate)
plt.xlabel('Number of customer arriving in a min')
plt.ylabel('Probability of customer arriving in a min')
plt.show()
```
