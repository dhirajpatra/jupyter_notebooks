```python
import matplotlib.pyplot as plt
import numpy as np

# Function to minimize
def f(x):
    return x**2  

# Derivative  
def df(x):
    return 2*x
  

# Gradient descent
learning_rate = 0.1  
x = 5 # Initial value

# Track values for plotting  
x_history = [x]

for i in range(10):
    x -= learning_rate * df(x)  
    # Track values
    x_history.append(x)
    print(f"Iteration {i+1}: x = {x}")

print(f"Final x is {x}")

# Plot optimization path
plt.plot(x_history)
plt.xlabel("Iteration") 
plt.ylabel("x value")
plt.title("Gradient Descent Optimization Path")

plt.show()
```
