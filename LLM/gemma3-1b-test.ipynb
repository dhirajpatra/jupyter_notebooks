## Testing OLLama running locally

```python
import ollama
```

```python
model_name = "gemma3:1b" # Or "gemma:4b"

prompt = "Explain the concept of black holes in simple terms."

response = ollama.generate(model=model_name, prompt=prompt)
print(response['response'])
```

```python
model_name = "gemma3:1b" # Or "gemma:4b"

messages = [
    {'role': 'user', 'content': 'What are the main benefits of renewable energy?'},
]

response = ollama.chat(model=model_name, messages=messages)
print(response['message']['content'])

# Continue the conversation
messages.append({'role': 'assistant', 'content': response['message']['content']})
messages.append({'role': 'user', 'content': 'Can you list some examples of renewable energy sources?'})

response = ollama.chat(model=model_name, messages=messages)
print(response['message']['content'])
```
