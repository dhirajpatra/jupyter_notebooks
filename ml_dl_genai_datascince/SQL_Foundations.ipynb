# Docker



### Installing Docker on Windows

1. **Download Docker Desktop for Windows:**
   - Go to the [Docker Desktop for Windows](https://www.docker.com/products/docker-desktop) download page.
   - Click on "Download Docker Desktop for Windows."

2. **Install Docker Desktop:**
   - Run the downloaded installer.
   - Follow the installation instructions in the setup wizard. Make sure to enable the "Use WSL 2 instead of Hyper-V" option if prompted, as this provides better performance.
   - After the installation, Docker Desktop will start automatically.

3. **Verify Installation:**
   - Open a terminal or command prompt and run:

     ```sh
     docker --version
     ```

   - You should see the version of Docker installed.

### Installing Docker on macOS

1. **Download Docker Desktop for Mac:**
   - Go to the [Docker Desktop for Mac](https://www.docker.com/products/docker-desktop) download page.
   - Click on "Download Docker Desktop for Mac."

2. **Install Docker Desktop:**
   - Open the downloaded `.dmg` file and drag the Docker icon to your Applications folder.
   - Open Docker from the Applications folder. You may be prompted to authorize Docker with your system password.

3. **Verify Installation:**
   - Open a terminal and run:

     ```sh
     docker --version
     ```

   - You should see the version of Docker installed.

### Installing Docker on Linux

The steps for installing Docker on Linux can vary slightly depending on the distribution. Here are the instructions for Ubuntu, one of the most common distributions:

1. **Update Your System:**

   ```sh
   sudo apt update
   sudo apt upgrade
   ```

2. **Install Required Packages:**

   ```sh
   sudo apt install apt-transport-https ca-certificates curl software-properties-common
   ```

3. **Add Dockers Official GPG Key:**

   ```sh
   curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo gpg --dearmor -o /usr/share/keyrings/docker-archive-keyring.gpg
   ```

4. **Add Docker Repository:**

   ```sh
   echo "deb [arch=amd64 signed-by=/usr/share/keyrings/docker-archive-keyring.gpg] https://download.docker.com/linux/ubuntu $(lsb_release -cs) stable" | sudo tee /etc/apt/sources.list.d/docker.list > /dev/null
   ```

5. **Install Docker:**

   ```sh
   sudo apt update
   sudo apt install docker-ce
   ```

6. **Start and Enable Docker:**

   ```sh
   sudo systemctl start docker
   sudo systemctl enable docker
   ```

7. **Verify Installation:**

   ```sh
   docker --version
   ```

   - You should see the version of Docker installed.


By following these steps, you should be able to install Docker on your respective operating system and verify that it is running correctly.

# Postgres on Docker

Running PostgreSQL in a Docker container is a common practice for development and testing environments. Below are the steps to get PostgreSQL up and running in Docker.

### Steps to Run PostgreSQL in Docker

1. **Pull the PostgreSQL Docker Image**

   First, you need to pull the PostgreSQL image from Docker Hub. Open your terminal and run the following command:

   ```sh
   docker pull postgres
   ```

2. **Run a PostgreSQL Container**

   After pulling the image, you can start a container using the following command:

   ```sh
   docker run --name my_postgres -e POSTGRES_PASSWORD=mysecretpassword -d postgres
   ```

   - `--name my_postgres`: Names the container `my_postgres`.
   - `-e POSTGRES_PASSWORD=mysecretpassword`: Sets the PostgreSQL password to `mysecretpassword`.
   - `-d`: Runs the container in detached mode.
   - `postgres`: The name of the image to use.

3. **Connect to PostgreSQL**

   To connect to the PostgreSQL instance running inside the Docker container, you can use the `psql` command-line tool or any PostgreSQL client.

   **Using psql:**

   ```sh
   docker exec -it my_postgres psql -U postgres
   ```

   **Using a PostgreSQL client:**

   - Host: `localhost`
   - Port: `5432`
   - Username: `postgres`
   - Password: `mysecretpassword`

4. **Persisting Data**

   By default, data stored in the PostgreSQL container will be lost when the container is removed. To persist data, you need to mount a volume.

   ```sh
   docker run --name my_postgres -e POSTGRES_PASSWORD=mysecretpassword -v my_postgres_data:/var/lib/postgresql/data -d postgres
   ```

   - `-v my_postgres_data:/var/lib/postgresql/data`: Mounts a volume named `my_postgres_data` to persist the database data.
   ```

5. **Stopping and Removing the Container**

   To stop the container:

   ```sh
   docker stop my_postgres
   ```

   To remove the container:

   ```sh
   docker rm my_postgres
   ```


By following these steps, you should be able to run PostgreSQL in Docker with ease.

# PRIMARY KEY & FOREIGN KEY
A primary key and a foreign key are essential concepts in relational databases that help maintain the integrity and establish relationships between tables.

### Primary Key
A primary key is a unique identifier for each record in a database table. It ensures that each record can be uniquely identified and accessed. A primary key has the following characteristics:
- **Uniqueness**: Each value in the primary key column must be unique.
- **Non-nullable**: The primary key column cannot have null values.
- **Immutable**: The value of a primary key should not change once assigned.

### Foreign Key
A foreign key is a column or a set of columns in one table that refers to the primary key in another table. The purpose of the foreign key is to establish and enforce a link between the data in the two tables. A foreign key has the following characteristics:
- **Referential Integrity**: Ensures that a value in the foreign key column must match an existing value in the primary key column of the referenced table.
- **Relationships**: Defines relationships between tables, such as one-to-many or many-to-many.

# psycopg2
Finally let's understand how to use Postgres via Python
### Setting Up the Environment

First, make sure you have PostgreSQL installed and running on your system. You also need to have Python and `psycopg2` installed. You can install `psycopg2` using pip:

```bash
pip install psycopg2
```

### Establishing a Connection

To interact with a PostgreSQL database, you need to establish a connection.

```python
import psycopg2

# Establish a connection to the database
try:
    connection = psycopg2.connect(
        user="your_username",
        password="your_password",
        host="127.0.0.1",
        port="5432",
        database="your_database"
    )
    cursor = connection.cursor()
    print("Connected to the database")
except (Exception, psycopg2.Error) as error:
    print("Error while connecting to PostgreSQL", error)
```

### Creating a Table

Heres how to create a table in PostgreSQL using Python:

```python
create_table_query = '''
CREATE TABLE IF NOT EXISTS employees (
    id SERIAL PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    age INT,
    department VARCHAR(50)
)
'''

try:
    cursor.execute(create_table_query)
    connection.commit()
    print("Table created successfully")
except (Exception, psycopg2.Error) as error:
    print("Error while creating table", error)
```

### Inserting Data

Inserting data into the table is like adding a new row to your spreadsheet.

```python
insert_query = '''
INSERT INTO employees (name, age, department)
VALUES (%s, %s, %s)
RETURNING id;
'''

values = ("John Doe", 30, "Engineering")

try:
    cursor.execute(insert_query, values)
    employee_id = cursor.fetchone()[0]
    connection.commit()
    print(f"Inserted data with id {employee_id}")
except (Exception, psycopg2.Error) as error:
    print("Error while inserting data", error)
```

### Reading Data

Reading data is like looking at rows in your spreadsheet. You can fetch all rows or specific ones based on certain criteria.

```python
select_query = '''
SELECT * FROM employees WHERE department = %s
'''
department = ("Engineering",)

try:
    cursor.execute(select_query, department)
    rows = cursor.fetchall()
    for row in rows:
        print(f"ID = {row[0]}, Name = {row[1]}, Age = {row[2]}, Department = {row[3]}")
except (Exception, psycopg2.Error) as error:
    print("Error while fetching data", error)
```

### Updating Data

Updating data is like editing a cell in your spreadsheet.

```python
update_query = '''
UPDATE employees SET age = %s WHERE id = %s
'''
values = (31, employee_id)

try:
    cursor.execute(update_query, values)
    connection.commit()
    print(f"Updated employee id {employee_id}")
except (Exception, psycopg2.Error) as error:
    print("Error while updating data", error)
```

### Deleting Data

Deleting data is like removing a row from your spreadsheet.

```python
delete_query = '''
DELETE FROM employees WHERE id = %s
'''
employee_to_delete = (employee_id,)

try:
    cursor.execute(delete_query, employee_to_delete)
    connection.commit()
    print(f"Deleted employee id {employee_id}")
except (Exception, psycopg2.Error) as error:
    print("Error while deleting data", error)
```

### Closing the Connection

Once you're done, it's important to close the connection to free up resources.

```python
if connection:
    cursor.close()
    connection.close()
    print("PostgreSQL connection is closed")
```

### Putting It All Together

Here's a complete example script that includes all the operations:

```python
import psycopg2

try:
    # Connect to your postgres DB
    connection = psycopg2.connect(
        user="your_username",
        password="your_password",
        host="127.0.0.1",
        port="5432",
        database="your_database"
    )
    cursor = connection.cursor()
    print("Connected to the database")

    # Create a table
    create_table_query = '''
    CREATE TABLE IF NOT EXISTS employees (
        id SERIAL PRIMARY KEY,
        name VARCHAR(100) NOT NULL,
        age INT,
        department VARCHAR(50)
    )
    '''
    cursor.execute(create_table_query)
    connection.commit()
    print("Table created successfully")

    # Insert data
    insert_query = '''
    INSERT INTO employees (name, age, department)
    VALUES (%s, %s, %s)
    RETURNING id;
    '''
    values = ("John Doe", 30, "Engineering")
    cursor.execute(insert_query, values)
    employee_id = cursor.fetchone()[0]
    connection.commit()
    print(f"Inserted data with id {employee_id}")

    # Read data
    select_query = '''
    SELECT * FROM employees WHERE department = %s
    '''
    department = ("Engineering",)
    cursor.execute(select_query, department)
    rows = cursor.fetchall()
    for row in rows:
        print(f"ID = {row[0]}, Name = {row[1]}, Age = {row[2]}, Department = {row[3]}")

    # Update data
    update_query = '''
    UPDATE employees SET age = %s WHERE id = %s
    '''
    values = (31, employee_id)
    cursor.execute(update_query, values)
    connection.commit()
    print(f"Updated employee id {employee_id}")

    # Delete data
    delete_query = '''
    DELETE FROM employees WHERE id = %s
    '''
    employee_to_delete = (employee_id,)
    cursor.execute(delete_query, employee_to_delete)
    connection.commit()
    print(f"Deleted employee id {employee_id}")

except (Exception, psycopg2.Error) as error:
    print("Error while connecting to PostgreSQL", error)

finally:
    if connection:
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
```
