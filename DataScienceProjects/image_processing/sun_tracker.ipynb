```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the input image
filename = 'images/sun_on_scene2.jpeg'
image = cv2.imread(filename)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
radius = 15
```

```python
# Find the coordinates of the brightest pixel
gray = cv2.GaussianBlur(gray, (radius, radius), 0)
(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(gray)
```

```python
# Draw a circle around the brightest spot
cv2.circle(image, maxLoc, radius, (0, 255, 0), 2)
```

```python
# Display the original image with circles around the bright spots
plt.imshow(image)
plt.title('Bright Spot Detection')
plt.show()
```
