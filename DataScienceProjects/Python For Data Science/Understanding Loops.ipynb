```python
#While loops in python
i = 1 #initiliza a variable i with 1

#Running a while loop
while i < 10:
    print(i)
    i += 1
```

```python
#For loops in python
color = ["Red", "Green", "Blue"]
for x in color:
    print(len(x))
```

```python
#A simple use case of loop

student_marks = [23,32,26,14] #marks obtained by students by class out of 20

percentage_marks = [] #creating an empty list

for i in student_marks:
    percentage_marks.append((i/80)*100)
    
percentage_marks
```
