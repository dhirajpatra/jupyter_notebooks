```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create a sample dataframe
df = pd.DataFrame({'car_id': [1, 2, 3, 4],
                   'fuel_type': ['gasoline', 'diesel', 'electric', 'gasoline']})

# create a mapping dictionary
fuel_type_mapping = {'gasoline': 0, 'diesel': 1, 'electric': 2}

# map the fuel_type column using the mapping dictionary
df['fuel_type_encoded'] = df['fuel_type'].map(fuel_type_mapping)

print(df)
```

```python
iris = datasets.load_iris()
clf = SVC()
clf.fit(iris.data, iris.target)
list(clf.predict(iris.data[:3]))
clf.fit(iris.data, iris.target_names[iris.target])
list(clf.predict(iris.data[:3]))
```

```python
x = np.arange(20).reshape(-1, 1)
y = np.array([5, 12, 11, 19, 30, 29, 23, 40, 51, 54, 74, 62, 68, 73, 89, 84, 89, 101, 99, 106])
x_train, x_test, y_train, y_test = train_test_split(x, y)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
model = LinearRegression().fit(x_train, y_train)
model.intercept_
model.coef_
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# Plot the data points and the regression line
plt.scatter(x, y, color='black')
plt.plot(x, model.predict(x), color='blue')
plt.show()
```

```python
sns.regplot(x=x, y=y, data=iris.data, ci = None)
plt.show()
```

```python
sns.residplot(x=x, y=y, data=iris.data, lowess=True)
plt.show()
```
