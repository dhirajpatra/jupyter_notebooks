### 1. Why Do We Need Data Visualization? Explain with Examples 

Data visualization is the process of converting data into a graphical format, such as charts, graphs, maps, or diagrams, in order to make it easier to understand and interpret. There are several reasons why data visualization is important:

Clarity: Data visualization can make complex data sets more understandable and accessible to a wider audience. For example, a chart or graph can quickly convey the overall trend or pattern in a set of data, while a map can show the geographic distribution of data.

Insight: Data visualization can reveal insights and patterns in data that might not be immediately apparent from raw numbers. For example, a graph that shows the relationship between two variables can reveal a correlation or causal relationship.

Communication: Data visualization can be used to communicate the results of data analysis to others, such as management, stakeholders, or the general public. A well-designed chart or graph can help to make a point or tell a story more effectively than raw numbers.

Decision Making: Data visualization can aid decision making by highlighting key trends, patterns and outliers in data. This can help people understand what is happening and what actions need to be taken.

Time and Cost-Effective: Data visualization allows for easy interpretation of large data sets, which can save time and resources.

Overall, data visualization makes data more accessible, understandable, and actionable, which can lead to better decision making and improved performance.

### 2. Give a Brief Note on Libraries used for Data Visualization. 

There are many libraries and frameworks available for creating data visualizations, each with their own strengths and use cases. Some popular libraries for data visualization include:

__Matplotlib__: A widely-used library for creating static, 2D plots in Python. It is highly customizable and can be used to create a wide range of plots, including line charts, bar charts, scatter plots, and histograms.

__Seaborn__: A library built on top of Matplotlib that provides a higher-level interface for creating more complex, visually appealing plots. It is particularly useful for creating statistical visualizations, such as heatmaps and violin plots.

__Plotly and Dash__: Libraries for creating interactive, web-based visualizations. They are built on JavaScript and can be used to create a wide range of plots, including bar charts, scatter plots, and maps.

__ggplot2__: A popular data visualization library for the R programming language. It is based on the Grammar of Graphics, a powerful and flexible approach to creating visualizations.

__D3.js__: A JavaScript library for creating interactive, web-based visualizations. It allows for the manipulation of documents based on data, and can be used to create a wide range of custom visualizations.

__Bokeh__: A Python library for creating interactive, web-based visualizations. It is built on JavaScript and can be used to create a wide range of plots, including line charts, bar charts, and maps.

__Tableau__: A commercial data visualization software that can be used to create interactive, web-based visualizations. It is widely used in businesses and organizations for data analysis and reporting.

These are just a few examples of the many libraries and frameworks available for data visualization. The best library to use depends on the specific requirements of the project and the skills of the developer

```python
def histogram(x, y, title, legend, x_label, y_label, width, height, n_bins):
    # Creating histogram
    fig, axs = plt.subplots(1, 1, figsize =(width, height), tight_layout = True)


    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        axs.spines[s].set_visible(False)

    # Remove x, y ticks
    axs.xaxis.set_ticks_position('none')
    axs.yaxis.set_ticks_position('none')

    # Add padding between axes and labels
    axs.xaxis.set_tick_params(pad = 5)
    axs.yaxis.set_tick_params(pad = 10)

    # Add x, y gridlines
    axs.grid(b = True, color ='grey',
            linestyle ='-.', linewidth = 0.5,
            alpha = 0.6)

    # Add Text watermark
    fig.text(0.9, 0.15, 'Jeeteshgavande30',
             fontsize = 12,
             color ='red',
             ha ='right',
             va ='bottom',
             alpha = 0.7)

    # Creating histogram
    N, bins, patches = axs.hist(x, bins = n_bins)

    # Setting color
    fracs = ((N**(1 / 5)) / N.max())
    norm = colors.Normalize(fracs.min(), fracs.max())

    for thisfrac, thispatch in zip(fracs, patches):
        color = plt.cm.viridis(norm(thisfrac))
        thispatch.set_facecolor(color)

    # Adding extra features   
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(legend)
    plt.title(title)

    # Show plot
    plt.show()
```

```python
# Creating dataset
np.random.seed(23685752)
N_points = 10000
n_bins = 20
 
# Creating distribution
x = np.random.randn(N_points)
y = .8 ** x + np.random.randn(10000) + 25
legend = ['distribution']
 
# calling the histogram function
histogram(x, y, 'Customized histogram', legend, "X-axis", "Y-axis", 10, 7, n_bins)
```
