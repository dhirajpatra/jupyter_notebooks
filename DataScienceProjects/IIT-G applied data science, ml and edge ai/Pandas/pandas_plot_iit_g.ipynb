```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 10))
 
# creating bar plot
plt.bar(courses, values, color ='BLUE',
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
Course_IT = [12, 30, 1, 8, 22]
Course_ECE = [28, 6, 16, 5, 10]
Course_CSE = [29, 3, 24, 25, 17]
 
# Set position of bar on X axis
br1 = np.arange(len(Course_IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make plots
plt.bar(br1, Course_IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT')
plt.bar(br2, Course_ECE, color ='g', width = barWidth, edgecolor ='grey', label ='ECE')
plt.bar(br3, Course_CSE, color ='b', width = barWidth, edgecolor ='grey', label ='CSE')
 
# Adding Axis
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Course_IT))],
        ['2015', '2016', '2017', '2018', '2019'])
 
plt.legend()
plt.show()
```

```python
def f(x, Y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
x = np.linspace(-6, -6, 30)
y = np.linspace(-6, -6, 30)
x, y = np.meshgrid(x, y)
z = f(x,y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z, 50, cmap='binary')
ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)
plt.show()
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# by matplotlib
sinplot()
plt.show()
```

```python
# by seaborn
sb.set()
sb.set_style('whitegrid')
sinplot()
plt.show()
```
